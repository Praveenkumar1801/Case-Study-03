````python
conditions = [
    (df5['Lifetime Giving'] == 0),
    (df5['Lifetime Giving'] > 0 )]

##Assigning value to the above condition

values = [0,1]

df5['Donation_made_or_not'] = np.select(conditions, values)
df5
````

````python
##Classtifying people
conditions = [
    (df5['Lifetime Giving'] >= 1000000),
    (df5['Lifetime Giving'] < 1000000) & (df5['Lifetime Giving'] >= 30000),
    (df5['Lifetime Giving'] < 30000) &(df5['Lifetime Giving'] > 0),
    (df5['Lifetime Giving'] == 0)]

##Assigning value to the above condition

values = [1,2,3,4]

df5['Category_based_on_given'] = np.select(conditions, values)
df5
````

````python
df7.info()
````

````python
## dropping one column that we created for encoding 

df8 = df7.drop(['Title_President','Preferred Language_E','Gender_M','Relationship Types_Graduate, Current Parent','Faculty of Graduation_Science -- MSE','Province_ON','Past Traveler Y/N_N'],axis='columns')
df8.info()
````

````python
from numpy.ma.core import shape
# Import DecisionTreeClassifier
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error
# Import train_test_split
from sklearn.model_selection import train_test_split
# Import accuracy_score
from sklearn.metrics import accuracy_score

## Lets create X and Y
X = df8.drop(columns=['Year of Last Gift','Year of First Gift','Donation_made_or_not','Category_based_on_given','Lifetime Giving']) #Since all these 5 columns are highly correlated with the outcome
y = df8.Donation_made_or_not
print(X.shape)
print(y.shape)
````

````python
# Split the dataset into 80% train, 20% test
X_train, X_test, y_train, y_test= train_test_split(X, y,test_size=0.2,random_state=1)
# INITIATE KNN
knn = KNeighborsClassifier(n_neighbors=15)
````

````python
# Fit dt to the training set
knn.fit(X_train,y_train)

# Predict the test set labels
y_pred = knn.predict(X_test)
# Evaluate the test-set accuracy
classification_report = classification_report(y_test,y_pred)
print(classification_report)
confusion_matrix = confusion_matrix(y_test,y_pred)
print(confusion_matrix)
mean_squared_error = mean_squared_error(y_test,y_pred)
print('Mean square vaue for this model is',mean_squared_error)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
````
